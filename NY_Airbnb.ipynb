```python
# Load subways dataset

df_subway = pd.read_csv('data/rawData/nyc-transit-subway-entrance-and-exit-data.csv')

df_subway.head()
```

```python
df_subway.info()
```

```python
station_locs = df_subway['Station Location'].apply(lambda x : list(json.loads(x.replace("'",'"')).values()) )
station_locs = np.array(station_locs.tolist())
station_locs = station_locs.astype(float)
```

```python
# Find nearest distance to subway station

import geopy.distance

def find_nearest(x):
    # approximate radius of earth in km
    R = 6373.0
    
#    #get lat and loc of airbnb
    lats0 = x['latitude']
    long0 = x['longitude']
    
    distance = list()
    
    #convert location to radians
    loc0 = np.radians(np.array([long0,lats0]).reshape(-1,2))
    
#     diff in long and lats
    dist = np.radians(station_locs) - loc0
    
#     diff in lats
    lats_diff = dist[:,1]
    
#     diff in long
    long_diff = dist[:,0]
    
#     calculate distance
    a = (np.sin(lats_diff/2))**2 + np.cos(lats0) * np.cos(station_locs[:,1]) * (np.sin(long_diff /2))**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1-a))
    distance = R * c
    
#     return minimum distancs in km
    return min(distance)
```

```python
# Get the distance to the nearest subway station

df_new_york["distance_to_nearest_subway"] = df_new_york.apply(lambda x : find_nearest(x),axis=1 )
df_new_york["distance_to_nearest_subway"]
```

```python
# Load crime dataset

df_crime = pd.read_csv('data/rawData/NYPD_Complaint_Data_Historic.csv', low_memory=False)
```

```python
# Search for complaints in 2019
df_crime['CMPLNT_FR_DT'] = pd.to_datetime(df_crime['CMPLNT_FR_DT'], errors = 'coerce')

# Use only 2019 data
df_crime = df_crime[df_crime['CMPLNT_FR_DT'].dt.year == 2019]
```

```python
df_crime.head()
```

```python
# Crimes by neighborhood

complaints_by_neighborhood = df_crime.groupby('BORO_NM').count()['CMPLNT_NUM'].sort_values(ascending=False)

complaints_by_neighborhood
```

```python
# Merge crime data with airbnb data

df_new_york['crimes'] = df_new_york["neighbourhood_group"].apply(lambda x: complaints_by_neighborhood[x.upper()])
```
